## Desafio 1 - Empréstimo

Em um financiamento com juros compostos e número de parcelas fixas parte-se dos seguintes parâmetros:
* `S` - valor da primeira parcela
* `N` - número de parcelas
* `J` - percentual de juros mensal

A primeira parcela a ser paga do financiamento é sempre igual a `S`. A partir daí é feita uma atualização mensal da parcela, em que cada nova parcela é calculada a partir da parcela do mês anterior, conforme a fórmula:

> Parcela<sub>mês</sub> = Parcela<sub>mês-1</sub> * (1 + `J` / 100)

O financiamento encerra quando as `N` parcelas são pagas.

Exemplo:
* `S`: 200
* `N`: 5
* `J`: 1%

Parcelas do financiamento:
`200`; `202`; `204.02`; `206.06`; `208.12`

### Desafio 1 / Parte 1 - Escrevendo o programa sem módulos

Dado o problema descrito, escreva um programa que calcule as parcelas de um empréstimo para os seguintes valores:
* S: 200
* N: 5
* J: 1%

Nesta versão não use outro módulo além da função principal.

```java
float parcela = 200;
int numParcelas = 5;
float juros = 1;

// Calculamos os valores das parcelas
for(int i = 1; i <= numParcelas; i++){
    System.out.println("Valor da parcela " + i + ": " + parcela);
    parcela = parcela * (1 + juros/100);
}
```

### Desafio 1 / Parte 2 - Escrevendo módulos

Reescreva o código acima de forma que seu programa faça uso de uma função que seja responsável pelo cálculo de uma parcela X do empréstimo. Podem ser usadas mais funções (métodos) conforme a necessidade.

```java
public class Emprestimo {
    
    static float calculaParcela(float valInicial, float juros, int numParcela){
        
        if(numParcela == 1){
            return valInicial;
        }
            
        float parcela = valInicial;
        juros /= 100;
        
        for(int i = 1; i < numParcela; i++){
            parcela = parcela * (1 + juros);
        }
        
        return parcela;
    }
}
```

```java
float valInicial = 200;
float juros = 1;
int numParcelaDesejada = 4;

float parcela = Emprestimo.calculaParcela(valInicial, juros, numParcelaDesejada);

System.out.println("Valor da parcela " + numParcelaDesejada + ": " + parcela);
```

## Desafio 2

Escreva um programa que sorteie um número inteiro entre `0` e `9999` o número deve ser mostrado, em seguida convertido para binário e a versão binária deve ser apresentada no console. A conversão decimal/binário deve ser computada por um programa feito por você.

```java
import java.util.Random;
public class randomInt {
    
    static int gerarRandomInt(int max){
        Random random = new Random();
        
        return random.nextInt(max);
    }
        
}
```

```java
import java.lang.Math;
public class binario {
    
    static long converteBinario(int numero){
        
        // Vetor para salvar cada bit do número em binário
        int[] numBinario = new int[32];
        
        // Conversão para binário
        int i = 0;
        while (numero > 0) {
            numBinario[i] = numero % 2;
            numero = numero / 2;
            i++;
        }
        
        // Conversão do numero em binário para uma única variável
        long binario = 0;
        for(int j = 0; j < i; j++){
            binario += numBinario[j] * Math.pow(10, j);
        }
        
        return binario;        
    }
    
}
```

```java
// Obtemos um número aleatório
int numero = randomInt.gerarRandomInt(9999);
System.out.println("Numero sorteado: " + numero);

// Convertemos o número para sua representação binária
long numeroBinario = binario.converteBinario(numero);
System.out.println("Representação em binario: " + numeroBinario);
```

## Desafio 3

Uma molécula de DNA pode ser definida a partir de uma cadeia que representa a sequência de suas bases:
A - Adenina
C - Citisina
G - Guanina
T - Tinina

Desse modo, uma string pode ser usada para representar um segmento do DNA da seguinte maneira: `ATTACGCGCAAAC`.

Escreva uma função (método) que codifique a cadeia de RNA produzida a partir de uma cadeia de DNA. A função deve ser genérica o suficiente para ser aplicável a qualquer cadeia. A entrada é uma string (DNA) e o retorno é uma string (RNA).

Escreva um programa que teste esta função com uma cadeia de DNA de sua escolha.

```java
public class DNA {
            
    static String converteRna(String dna){
        
        // Variável para armazenar cadeia de RNA
        String rna = "";
        
        for(int i = 0; i < dna.length(); i++){
            // Convertemos a base do DNA para sua versão RNA
            char baseRna = dna.charAt(i);

            if(baseRna == 'A'){
                baseRna = 'U';
            }

            if(baseRna == 'G'){
                baseRna = 'C';
            }

            if(baseRna == 'C'){
                baseRna = 'G';
            }

            if(baseRna == 'T'){
                baseRna = 'A';
            }
            
            // Adicionamos a nova base à cadeia de RNA
            rna = rna + baseRna;
        }
        
        return rna;        
    }
      
}    
```

```java
String cadeiaOrig = "ATTACGCGCAAAC";
String cadeiaSaida = DNA.converteRna(cadeiaOrig);

System.out.println("Cadeia de DNA: " + cadeiaOrig);
System.out.println("Cadeia de RNA transcrita: " + cadeiaSaida);
```

## Desafio 4

Uma empresa precisa realizar uma estatística do salário de seus funcionários.

Para fins de teste, os salários devem ser gerados aleatoriamente com valores variando entre R\\$ 2.500 e R\\$ 15.000.

Escreva um programa que gere uma lista contendo os salários de 50 funcionários da empresa e mostre no console quantos funcionários ganham salário acima da média.

```java
import java.util.Random;
public class Helpers {
    
    public static float gerarRandomFloat(float min, float max){
        Random r = new Random();
        
        return min + r.nextFloat() * (max - min);
    }
    
        
    public static float calculaMedia(float[] array){
        float media = 0;
        int tam = array.length;
        
        for(int i = 0; i < tam; i++){
            media += array[i]/tam;
        }
        
        return media;
    }
        
}
```

```java
public class funcionarios {
    
    public static float[] geraSalarios(int numFuncionarios, float minSalario, float maxSalario){
        
        // Array que armazena o salário de cada funcionário
        float[] salarios = new float[numFuncionarios];
        
        for(int i = 0; i < numFuncionarios; i++){
            salarios[i] = Helpers.gerarRandomFloat(minSalario, maxSalario);
        }
        
        return salarios;
    }
    
    public static int totalAcimaMedia(float[] salarios){
        // Calculamos a média dos salários
        float media = Helpers.calculaMedia(salarios);
        
        // Contamos quantos salários estão acima da média calculada
        int totalAcima = 0;
        for(int i = 0; i < salarios.length; i++){
            if(salarios[i] > media){
                totalAcima++;
            }
        }
        
        return totalAcima;
    }
    
}
```

```java
float[] salarios = funcionarios.geraSalarios(50, 2500, 15000);

System.out.println("Lista de salarios:");
for(int i = 0; i < salarios.length; i++){
    System.out.println(salarios[i]);
}

int acimaMedia = funcionarios.totalAcimaMedia(salarios);

System.out.println(acimaMedia + " funcionario(s) recebe(m) salário(s) acima da média");
```
